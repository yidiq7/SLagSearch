```python
X = sp.symbols('x0:5', real=True)
Y = sp.symbols('y0:5', real=True)
XY = X + Y

substitutions = {z: x + sp.I*y for z, x, y in zip(Z, X, Y)}

# 4. Perform the substitution
f_substituted = f.subs(substitutions)
cy_real, cy_imag = f_substituted.expand(complex=True).as_real_imag()
```

```python
imag_basis_z = []
for i in range(5):
    for j in range(i + 1, 5):
        expr = sp.im(Z[i] * sp.conjugate(Z[j]))
        imag_basis_z.append(expr)

# Next 15 basis functions: Re(z_i * conjugate(z_j)) for i <= j
real_basis_z = []
for i in range(5):
    for j in range(i, 5):
        expr = sp.re(Z[i] * sp.conjugate(Z[j]))
        real_basis_z.append(expr)

# The complete basis in terms of z
basis_z = imag_basis_z + real_basis_z

print(f"Successfully generated {len(basis_z)} basis functions.")
print("Example imaginary basis function (i=0, j=1):")
sp.pprint(imag_basis_z[0])
print("\nExample real basis function (i=0, j=0):")
sp.pprint(real_basis_z[0])
```

```python
# --- Convert the basis functions to x and y coordinates ---

# Define real symbols for x_k and y_k

# Create the substitution rule {z_k: x_k + I*y_k}
substitution_rule = {zk: xk + sp.I*yk for zk, xk, yk in zip(Z, X, Y)}

# Apply the substitution to each basis function and simplify
# The .doit() method evaluates the Re() and Im() functions
basis_xy = [b.subs(substitution_rule).doit() for b in basis_z]
```

```python
A = sp.symbols('a0:25', real=True)
B = sp.symbols('b0:25', real=True)
C = sp.symbols('c0:25', real=True)

# Create the full linear combination of the basis functions
# linear_combination = sum(wk * bk for wk, bk in zip(w, basis_xy))
eq1 = sp.Add(*[ak * basisk for ak, basisk in zip(A, basis_xy)])
eq2 = sp.Add(*[bk * basisk for bk, basisk in zip(B, basis_xy)])
eq3 = sp.Add(*[ck * basisk for ck, basisk in zip(C, basis_xy)])
```

```python
def get_restriction(self, ignored_coord=None, lambdify=False):
    if ignored_coord is None:
        ignored_coord = self.max_grad_coordinate
    # Since we have more than one ignored_coordinate in CICY, sympy subs()
    # cannot replace two coordinates simultaneously. As a result, if the first
    # expression contains the coordinate to be replaced by the second expression
    # that coordinate will also be replaced. So here we will create a temporary 
    # coordinate list W to avoid this issue
    W = sp.var('w0:{}'.format(len(self.affine_coordinates)))
    ignored_coordinate = np.array(W)[ignored_coord]
    local_coordinates = sp.Matrix(W).subs({coord: func for coord, func in zip(ignored_coordinate, self.function)})
    local_coordinates = local_coordinates.subs({w: z for w, z in zip(W, self.affine_coordinates)})
    restriction = local_coordinates.jacobian(self.affine_coordinates).inv()
    for coord in reversed(ignored_coord):
        restriction.col_del(coord)
    if lambdify is True:
        restriction = sp.lambdify([self.coordinates], restriction, 'numpy')
    return restriction
```
